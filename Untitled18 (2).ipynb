# HOLT WINTERS - Forecasting 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C://Users//Snigdha.Cheekoty//OneDrive - Serco//Desktop//practice//airline_passengers.csv", 
                 index_col = "Month",
                parse_dates = True)
```

```python
df.dropna(inplace = True)
```

```python
df.head(10)
```

```python
df.index.freq = "MS"
```

```python
df.index
```

```python
df.tail(10)
```

```python
len(df)
```

```python
## Train test split
## Test dataset should be the MOST RECENT DATA
train_data = df.iloc[:109]
test_data = df.iloc[108:]
```

```python
df.plot(figsize = (12,5))
```

```python
## From above plot: we decide to apply multiplicative model because: trend is increasing exponentially, seasonality is present
## Fitting the model
from statsmodels.tsa.holtwinters import ExponentialSmoothing
fitted_model = ExponentialSmoothing(train_data["Thousands of Passengers"], trend = "mul",seasonal = "mul", seasonal_periods = 12).fit()
```

```python
type(fitted_model)
```

```python
# Predicting on test dataset
test_predictions = fitted_model.forecast(12)
```

```python
test_predictions
```

```python
## Train- test split
train_data["Thousands of Passengers"].plot(legend = True, label = "Train", figsize = (12,5))
test_data["Thousands of Passengers"].plot(legend = True)
```

```python
# Plot the test predictions for 1 year
train_data["Thousands of Passengers"].plot(legend = True, label = "Train", figsize = (12,5))
test_data["Thousands of Passengers"].plot(legend = True)
test_predictions.plot(legend = True, label = "Test Predictions")
```

```python
# Zoomed-in Plot the test predictions for 1 year
df["Thousands of Passengers"].plot(legend = True, label = "Train", figsize = (12,5), xlim= ("1958-01-01", "1959-01-01"))
test_predictions.plot(legend = True, label = "Test Predictions", xlim = ("1958-01-01", "1959-01-01"))

```

```python
test_predictions = fitted_model.forecast(36)
```

```python
# Plot the test predictions for 3 years
train_data["Thousands of Passengers"].plot(legend = True, label = "Train", figsize = (12,5))
test_data["Thousands of Passengers"].plot(legend = True)
test_predictions.plot(legend = True, label = "Test Predictions")
```

```python
## Evaluating the model performance
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_squared_error(y_true = test_data, y_pred = test_predictions)
```

```python
mean_absolute_error(test_data, test_predictions)
```

```python
## Calculating RMSE value
np.sqrt(mean_squared_error(test_data, test_predictions))
```

```python
test_data.describe()
### Observe that RMSE is less than STD of the test data , thus model is performing well!!!
```

```python
# Forecasting into the future
final_model = ExponentialSmoothing(df["Thousands of Passengers"], trend = "mul", seasonal = "mul", seasonal_periods = 12).fit()
```

```python
forecasted_values = final_model.forecast(36)
```

```python
## Plotting the forecast into the future
df["Thousands of Passengers"].plot(figsize = (12,5), legend = True)
forecasted_values.plot(legend = True)
```

# ARIMA 

```python
## AUTO REGRESSIVE MODEL

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("C://Users//Snigdha.Cheekoty//OneDrive - Serco//Desktop//practice//uspopulation.csv", parse_dates = True,
                  index_col = "DATE")
```

```python
data.head()
```

```python
data.index.freq = "MS"
```

```python
data.index
```

```python
## Plotting the observations

data.plot(figsize = (12,8))
```

```python
len(data)
```

```python
## Split the data 

train_data = data.iloc[:84]
test_data = data.iloc[84:]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
## Fitting the model

from statsmodels.tsa.ar_model import AR, ARResults
model = AR(train_data["PopEst"])
```

```python
AR1fit = model.fit(maxlag = 1, method = "cmle")
```

```python
AR1fit.aic
```

```python
AR1fit.k_ar
```

```python
AR1fit.params
```

```python
start = len(train_data)
end = len(train_data) + len(test_data) - 1
```

```python
start
```

```python
end
```

```python
## Predictions
AR1fit.predict(start = start, end = end)
```

```python
## Comparing the above predicted values with actual test values

test_data
```

```python
predictions1 = AR1fit.predict(start = start, end = end)
predictions1 = predictions1.rename("AR(1) Predictions")
test_data.plot(figsize = (12,8), legend = True)
predictions1.plot(legend = True)


### Not so bad !! General trend is the same!
```

```python
### Auto regressive (AR model) of order 2

model = AR(data["PopEst"])
AR2fit = model.fit(maxlag = 2)
```

```python
AR2fit.params
```

```python
## predictions

predictions2 = AR2fit.predict(start = start, end = end)
predictions2 = predictions2.rename("AR(2) Predictions")
```

```python
## Plotting 

test_data.plot(figsize = (12,8), legend = True)
predictions2.plot(legend = True)


## Too good!!!
```

```python
### We don't need to apply the AR model in every order to check which is performing the best.
## Statsmodels does the work for us.

ARfit = model.fit(ic = "t-stat") 
## if we don't set the maxlag argument, then statsmodels chooses the best performing order for us!!

```

```python
ARfit.params

## Statsmodels decided that our model should have 12 lags in order to perform the best!
```

```python
predictions12 = ARfit.predict(start = start, end = end)
```

```python
predictions12 = predictions12.rename("AR(12) Predictions")
```

```python
from sklearn.metrics import mean_squared_error
labels = ["AR1", "AR2", "AR12"]
preds = [predictions1, predictions2, predictions12]
```

```python
for i in range(3):
    error = mean_squared_error(test_data["PopEst"], preds[i])
    print(labels[i] , "MSE error:" , error)
    
## AR12 has the least MSE Error! The order statsmodels chose for us is indeed correct! :) 
```

```python
## Plotting the AR12 predictions

test_data.plot(legend = True, figsize = (12,5))
predictions12.plot(legend = True)
```

```python
## Forecsting on future values 

model = AR(data["PopEst"])
ARfit = model.fit()
```

```python
ARfit.params

## statsmodels again chose 12th order for us! 
```

```python
start = len(data)
end = len(data) +12
forecasted_values = ARfit.predict(start  = start, end = end)
```

```python
forecasted_values
```

```python
forecasted_values = forecasted_values.rename("AR Forecast")
data.plot(figsize = (12,8), legend = True)
forecasted_values.plot(legend = True)
```
